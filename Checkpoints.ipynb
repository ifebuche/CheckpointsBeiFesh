```python
import pandas as pd
import re, folium
import matplotlib.pyplot as plt
```

```python
###################
#Send Out from here - This function will require the po
#####################
def mapping_fesh():
    import pandas as pd
    import re, folium
    import matplotlib.pyplot as plt

    print("This is a simple folium map with actual data by Fesh.")

    points = pd.read_csv("points.csv", header=None)
    points.head()
    points.rename(columns= {0:'lat', 1:'lon'}, inplace=True)
    lats = [round(line, 5) for line in points.lat]
    lons = [round(line, 5) for line in points.lon]


    fesh_map = folium.Map(location=(lats[0], lons[0]),
                            zoom_start=9,
                            radius=95,
                            color="#E37222",
                            fill=True, 
                         tiles = "openstreetmap")
    folium.TileLayer('stamenterrain', attr="<a href=https://deparkes.co.uk/2016/06/10/folium-map-tiles/>Endless Sky</a>").add_to(fesh_map)
    folium.TileLayer('cartodbpositron', attr="<a href=https://deparkes.co.uk/2016/06/10/folium-map-tiles/>Endless Sky</a>").add_to(fesh_map)
    folium.TileLayer('cartodbdark_matter', attr="<a href=https://deparkes.co.uk/2016/06/10/folium-map-tiles/>Endless Sky</a>").add_to(fesh_map)
    folium.LayerControl().add_to(fesh_map)

    for lat,lon, in zip(lats, lons):
        folium.Marker(location=[lat, lon], tooltip="Click for Google Maps link").add_to(fesh_map)

    #Show the map
    return fesh_map

fesh_map = mapping_fesh()
```
